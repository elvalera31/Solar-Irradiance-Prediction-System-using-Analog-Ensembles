                            Descripción: Este código automatiza el
                           filtrado de archivos CSV con datos de
                          estaciones: carga cada archivo, convierte
                          la columna de tiempo a formato datetime,
                          ordena los registros, los remuestrea a
                          intervalos de 3 minutos aplicando
                          promedios, elimina filas vacías y guarda
                          una versión filtrada y limpia de cada
                          archivo en la carpeta destino.

```python
import pandas as pd
import os
from google.colab import drive

# Montaje de Google Drive
drive.mount('/content/drive')

# Configuración de rutas y parametros
INPUT_FOLDER = '/content/drive/MyDrive/Datos Filtrados/'
OUTPUT_FOLDER = '/content/drive/MyDrive/Datos Filtrados/'
FILE_PATTERN = 'Copy of station_data_'
TIME_COLUMN = 'timestamp'
RESAMPLE_FREQ = '3min'

# Crear carpeta de salida si no existe
if not os.path.exists(OUTPUT_FOLDER):
    os.makedirs(OUTPUT_FOLDER)

def process_and_resample_file(input_file_path: str, output_file_path: str, time_col: str, freq: str) -> bool:
    """Carga, ordena por tiempo, remuestrea por promedio y guarda el resultado."""
    try:
        df = pd.read_csv(input_file_path)
        df[time_col] = pd.to_datetime(df[time_col])
        df = df.set_index(time_col).sort_index()
        df_resampled = df.resample(freq).mean().dropna(how='all')
        df_resampled.to_csv(output_file_path)
        print(f"Procesado y Guardado: {os.path.basename(output_file_path)}")
        return True
    except Exception as e:
        print(f"Error en {os.path.basename(input_file_path)}: {e}")
        return False

print(f"\nIniciando procesamiento en: {INPUT_FOLDER}")

# Iteracion y procesamiento de archivos
for filename in os.listdir(INPUT_FOLDER):
    if filename.startswith(FILE_PATTERN) and filename.endswith(".csv"):
        input_path = os.path.join(INPUT_FOLDER, filename)
        new_name_prefix = filename.replace('Copy of ', '').replace('.csv', '')
        output_filename = f"{new_name_prefix}_filtrado.csv"
        output_path = os.path.join(OUTPUT_FOLDER, output_filename)
        process_and_resample_file(input_path, output_path, TIME_COLUMN, RESAMPLE_FREQ)

print("\nProceso completado.")
```
